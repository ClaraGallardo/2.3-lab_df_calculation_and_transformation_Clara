```python
# your code here

pokemon = pd.read_csv('Pokemon.csv')

#El total es la suma de todos los stats?

pokemon
```

```python
stats = pokemon[['HP',	'Attack',	'Defense',	'Sp. Atk',	'Sp. Def',	'Speed']]
stats['total']= 0
stats
```

```python
stats['total']=stats[['HP',	'Attack',	'Defense',	'Sp. Atk',	'Sp. Def',	'Speed']].sum(axis=1)

stats
```

```python
stats['total']==pokemon['Total'] #asi vemos que la suma de los stats es el TOTAL¡
```
